```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
#Generate Data

np.random.seed(121)

N = 1000
std = 1

balloons = pd.DataFrame({
    'length': np.random.normal(30, std, N),
    'width' : np.random.normal(15, std, N),
    'weight' : np.random.normal(1, std, N),
    'class' : np.zeros(N, dtype='int')
})

melons = pd.DataFrame({
    'length' : np.random.normal(23, std, N),
    'width' : np.random.normal(12, std, N),
    'weight': np.random.normal(100, std, N),
    'class': np.ones(N, dtype='int')
})

mel_ball = pd.concat([balloons, melons])
```

```python
# Create the figure
fig = plt.figure()
ax = fig.add_subplot(111, projection='3d')

ax.scatter3D(xs=balloons['length'], ys=balloons['width'],
              zs=balloons['weight'], label = 'balloons',color='red')

ax.scatter3D(xs=melons['length'], ys=melons['width'],
             zs=melons['weight'], label = 'melons',color='yellow')

ax.set_xlabel("width")
ax.set_ylabel("length")
ax.set_zlabel("weight")

plt.show()
```

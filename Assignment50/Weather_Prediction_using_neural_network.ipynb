```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.model_selection import train_test_split
import tensorflow as tf
#from perceptron import Perceptron
from datetime import datetime
```

```python
df = pd.read_csv("data/weatherHistory.csv")
df.head(10)
```

```python
print('df shape before null drop:',df.shape)
df.isnull().sum()
```

```python
df = df.dropna()
print('df shape after null drop:', df.shape)
```

```python
#check data types of columns
print('Columns data types',df.dtypes)
```

```python
#extract date part from Formatted Date
df['Date_temp'] = df['Formatted Date'].str.strip().str[:10]
df.head()
```

```python
# Convert to datetime
df['Date_temp'] = pd.to_datetime(df['Date_temp'], format='%Y-%m-%d')
df.head()
```

```python
print(df.dtypes)
```

```python
# Format datetime values to '20060401' format
df['Date'] = df['Date_temp'].dt.strftime('%Y%m%d')
df.head()
```

```python
print(df.dtypes)
```

تاریخ های مختلف را تبدیل به «چندمین روز سال» نمایید. برای مثال ۲ اردیبهشت برابر است با ۳۳ امین روز سال، و ۲۹ اسفند برابر است با ۳۶۵ امین روز سال.



```python
df['day_of_year'] = df['Date_temp'].dt.dayofyear
#df = df.sort_values(by=['Date'], ascending=True).reset_index()
df.head()
```

از ۲۴ دمای مختلف که به ازای ساعت های هر روز در اختیار دارید، میانگین گیری نمایید. نتیجه نهایی شما باید مشابه جدول زیر باشد: دقت کنید که ممکن است در خروجی، دو سطر با مقدار ۳۶۵ وجود داشته باشد. برای مثال مقدار ۳۶۵ اول به ۲۹ اسفند سال ۸۶ اشاره می نماید. و ۳۶۵ دوم به ۲۹ اسفند سال ۸۹ اشاره می نماید

```python
# make column with this format ---> format='%m-%d %H'
df['MonthDayHour_datetime'] = pd.to_datetime(df['Formatted Date'], format='%Y-%m-%d %H')
```

```python
# make a 'MonthDayHour' column
df['MonthDayHour'] = df['Formatted Date'].str.strip().str[5:13]
df.head()
```

```python
# construct Day of years mean temperature (C) table
data = df.groupby(['Date','day_of_year'])['Temperature (C)'].mean().reset_index()
data
```

```python
data['Mean Temperature (C)'] = data['Temperature (C)']
```

```python
mean_temp_day_of_year = pd.concat([data['Mean Temperature (C)'], data['day_of_year']], axis=1)
mean_temp_day_of_year
```

```python
plt.figure(figsize=(24, 8))
plt.scatter(data['day_of_year'], data['Mean Temperature (C)'],
            marker='o',color='purple', linewidths=0.5, alpha=0.7)
plt.title('Day of years mean temperature (C)')
plt.xlabel('day of year')
plt.ylabel('Mean Temperature (C)')
plt.show()
```

```python
X = data['day_of_year'].values
Y= data['Mean Temperature (C)'].values
X.shape, Y.shape
```

```python
X = X.reshape(-1,1)
Y = Y.reshape(-1,1)
X.shape, Y.shape
```

```python
print('Type X:', type(X))
print('Type Y:', type(Y))
```

```python
X_train, X_test, Y_train, Y_test = train_test_split(X, Y, random_state=50, test_size=0.2)

X_train.shape, X_test.shape, Y_train.shape, Y_test.shape
```

```python
X_train.dtype, X_test.dtype, Y_train.dtype, Y_test.dtype
```

## Model

```python
from keras.layers.serialization import activation
model = tf.keras.models.Sequential([
    tf.keras.layers.Dense(32, activation="relu"),
    tf.keras.layers.Dense(16, activation="relu"),
    tf.keras.layers.Dense(8, activation="relu"),
    tf.keras.layers.Dense(1, activation="linear")
])
```

```python
model.compile(optimizer=tf.keras.optimizers.Adam(learning_rate=0.01),
    loss='mean_squared_logarithmic_error',
    metrics='RootMeanSquaredError')
```

```python
output = model.fit(X_train, Y_train, epochs=60, validation_data=(X_test, Y_test), validation_split=0.2)
```

```python
model_loss_history = pd.DataFrame({'loss': output.history['loss'],
                                   'val_loss': output.history['val_loss']})

model_loss_history.plot()
```

```python
model_predictions = model.predict(X_test)
model_predictions[20], Y_test[20]
```

```python
plt.figure(figsize=(24, 8))
plt.scatter(data['day_of_year'], data['Mean Temperature (C)'],
            marker='o',color='purple', linewidths=0.5, alpha=0.7)
plt.plot(X_test, model_predictions, ">",  color='red', lw=4)
plt.title('Day of years mean temperature (C)')
plt.xlabel('day of year')
plt.ylabel('Mean Temperature (C)')
plt.show()
```

```python
model.evaluate(X_test, Y_test)
```

```python
errors = 0
for index in range(len(Y_test)):
  if abs(model_predictions[index] - Y_test[index]) > 6:
    errors += 1

print(1 - (errors / len(Y_test)))
```

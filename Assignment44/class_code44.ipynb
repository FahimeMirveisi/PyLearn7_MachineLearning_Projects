```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
data = pd.read_csv('abalone.csv')
data.head()
```

```python
data["Sex"] = data["Sex"].replace(["F", "M", "I"], [0, 1, 2])
```

```python
data.corr()
```

```python
plt.scatter(data['Length'], data['Height'])
plt.show()
```

## LLS

```python
X_train = np.array(data["Length"])
Y_train = np.array(data["Height"])
```

```python
X_train = X_train.reshape(-1, 1)
```

```python
# w = شیب خط
# w = (X.T * X)^ -1 * X.T *Y
```

```python
w = np.matmul(np.matmul(np.linalg.inv(np.matmul(X_train.T , X_train)), X_train.T), Y_train)

#w = np.matmul((np.matmul((np.linalg.inv(np.matmul(X_train.T, X_train))), X_train.T)), Y_train)
```

```python
w
```

```python
plt.scatter(data['Length'], data['Height'])
plt.plot(X_train, w * X_train, color = 'red')
plt.plot(X_train, 0.87 * X_train, color = 'green')
plt.show()
```

```python
#predict

new_length = 0.49
x = new_length
y_pred = w * x

height = y_pred
print(height)
```

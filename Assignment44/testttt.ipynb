```python
x = x.reshape((-1, 1))
y = y.reshape((-1, 1))
print(x.shape)
print(y.shape)
```

```python
# w = inv(X.T @ X) @ X.T @ Y
m = np.matmul(np.matmul(inv(np.matmul(x.T , x)), x.T), y)
m
```
